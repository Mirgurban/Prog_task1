```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analit(x,y):
    return np.sin(np.pi * x) * np.sin(np.pi * y)
```

```python
N = 200
x_help = np.linspace(0, 1, N)
y_help = np.linspace(0, 1, N)
X, Y = np.meshgrid(x_help, y_help)
# print(np.meshgrid(x_help, y_help))
Z = analit(X, Y)
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```
